```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
mono = pd.read_csv('mono.csv')
mono
```

```python
dotnet2 = pd.read_csv('dotnet2.2.csv')
dotnet2
```

```python
dotnet3 = pd.read_csv('dotnet3.0.csv')
dotnet3
```

```python
df = pd.concat([mono, dotnet2, dotnet3])
df
```

```python
names = df['name'].unique()

for name in names:
    plt.clf()
    custom = df[df["name"]==name]
    custom = custom[custom["size"] >= 20]
    plt.title(name)
    chart = sns.pointplot(data=custom, x="size", y="ms", hue="domain")
    chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
    plt.savefig(f"{name}.png")
```
